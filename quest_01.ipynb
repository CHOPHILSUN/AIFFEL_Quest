print("뒤집힌 단어는", str(a), "입니다.") #str함수로 받은 문자열을 출력함.

if txt == a:          # 두 문자열을 비교해서 같으면 TURE 출력
  print("회문입니다")
else                  # 위 조건이 아닐 경우 FALSE 출력
  print("회문이 아닙니다")
